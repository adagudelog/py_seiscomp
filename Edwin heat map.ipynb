```python
# -*- coding: utf-8 -*-
import pandas as pd
import seaborn as sns
import matplotlib as pl
%matplotlib inline
```

```python
df = pd.read_csv("datoshistograma2.csv", index_col=0)
df = df.fillna(0)
```

```python

ax = sns.heatmap(df, cmap="YlGnBu")
ax.set_xlabel("Mes")
ax.set_ylabel(u"Año")
ax.set_title(u"Número de sismos registrados por la estación PRV")
f = ax.get_figure()
f.set_size_inches(11.7, 8.27)
f.savefig("n_simos_PRV.JPG")
```
